# 그래프

## 그래프란?

연결되어있는 원소간의 관계를 표현한 자료구조이고 연결할 객체를 나타내는 정점(Vertext)과 객체를 연결하는 간선(Edge)의 집합으로 구성된다. 그래프 G를 G=(V, E)로 정의하는데, V는 정점의 집합, E는 간선들의 집합을 의미한다.

ex) 지하철 노선도, KTX 노선도 도시 도로망, 전기 회로도, 인맥 관계도

## 그래프의 종류

- 무방향 그래프(Undirected Graph)
    - 간선에 방향성이 없는 그래프

- 방향 그래프(Directed Graph)
    - 화살표로 간선 방향을 표기하고, 그래프의 정점 집합이 무방향 그래프와 같음.

    image.png

- 가중치 그래프
    - 간전마다 가중치가 다르게 부여된 그래프
    - 무방향 그래프와 방향 그래프에 각각 가중치를 부여한 경우

    image.png

## 깊이 우선 탐색의 작동

- 그래프의 모든 정점을 한 번씩 방문하는 것을 그래프 순회(Graph Traversal)라고 함.
- 그래프 순회 방식은 깊이 우선 탐색(DFS), 너비 우선 탐색이 대표적(BFS).

- 인정 행렬은 정방형으로 구성된 행렬로 정점이 4개인 그래프는 4 x 4로 표현.

- 방향 그래프이 인접 행렬

## 그래프 구현

```python
class Graph():
    def __init__(self, size):
        self.SIZE = size
        self.graph = [[0 for _ in range(size)] for _ in range(size)]

G1 = None

G1 = Graph(4)

G1.graph[0][1] = 1  ## (A, B) 간선
G1.graph[0][2] = 1  ## (A, C) 간선
G1.graph[0][3] = 1  ## (A, D) 간선

G1.graph[1][0] = 1  ## (B, A) 간선
G1.graph[1][2] = 1  ## (B, C) 간선

G1.graph[2][0] = 1  ## (C, A) 간선
G1.graph[2][1] = 1  ## (C, B) 간선
G1.graph[2][3] = 1  ## (C, D) 간선

G1.graph[3][0] = 1  ## (D, A) 간선
G1.graph[3][2] = 1  ## (D, C) 간선

## G1 그래프 출력

print('## 무방향 그래프 ##')
for row in range(4):
    for col in range(4):
        print(G1.graph[row][col], end = '\t')
    print()
```

## 그래프의 구현2

```python
class Graph():
    def __init__(self, size):
        self.SIZE = size
        self.graph = [[0 for _ in range(size)] for _ in range(size)]

def printGragh(g, nameAry):
    print(' ', end = '\t')                                      ## end = ' '이 아니라 end = '\t'을 사용하면 출력되는 그래프의 가독성이 향상됨.
    for v in range(g.SIZE):
        print(nameAry[v], end = '\t')
    print()
    for row in range(g.SIZE):
        print(nameAry[row], end = '\t')
        for col in range(g.SIZE):
            print(g.graph[row][col], end = '\t')
        print()
    print()

G1 = None

nameAry = ['문별', '솔라', '휘인', '쯔위', '선미', '화사']
문별, 솔라, 휘인, 쯔위, 선미, 화사 = 0, 1, 2, 3, 4, 5           ## 변수 이름을 정점 번호로 지정하면 더 직관적

gsize = 6
G1 = Graph(gsize)

G1.graph[문별][솔라] = 1; G1.graph[문별][휘인] = 1; G1.graph[솔라][문별] = 1
G1.graph[솔라][쯔위] = 1; G1.graph[휘인][문별] = 1; G1.graph[휘인][쯔위] = 1
G1.graph[쯔위][솔라] = 1; G1.graph[쯔위][휘인] = 1; G1.graph[쯔위][선미] = 1
G1.graph[쯔위][화사] = 1; G1.graph[선미][쯔위] = 1; G1.graph[선미][화사] = 1
G1.graph[화사][쯔위] = 1; G1.graph[화사][선미] = 1
print(end = '\t')
print(end = '\t')

print('### G1 무방향 그래프 ###')
print()
printGragh(G1, nameAry)
```
